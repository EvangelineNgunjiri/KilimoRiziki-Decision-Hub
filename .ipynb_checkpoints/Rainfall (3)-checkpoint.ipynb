# Modeling

### Check stationarity

We shall test for stationarity using the ADfuller test

```python
from statsmodels.tsa.stattools import adfuller

# Assuming df_dict is a dictionary of DataFrames
for key, data in df_dict.items():
    result = adfuller(data['Value'])
    print(f'ADF Statistic for {key}: {result[0]}')
    print(f'p-value: {result[1]}')
    print(f'Critical Values: {result[4]}')

    if result[1] <= 0.05:
        print(f'The time series for {key} is likely stationary.')
    else:
        print(f'The time series for {key} is likely non-stationary.')
    print('-' * 40)
```

We shall create a difference dictionary consisting of the differenced data.

```python
df_dict_diff = {}
for key in df_dict.keys():
    # Assuming you want to calculate the difference for the 'Value' column
    #df_dict[key] = df_dict[key].set_index('Year').drop(columns='Region')
    df_dict_diff[key] = df_dict[key]['Value'].diff().dropna()


```

```python
import pandas as pd
import matplotlib.pyplot as plt
import pmdarima as pm
from statsmodels.tsa.arima.model import ARIMA

def arima_model(ref):
    # Perform grid search for ARIMA parameters using only the training data
    arma_model = pm.auto_arima(ref['Value'], seasonal=False, suppress_warnings=True)




    ref['Year'] = pd.to_datetime(ref['Year'])
    ref.set_index('Year', inplace=True)

    # ARIMA
    arima_order = arma_model.order  # Use different variable names
    arima_model_fit = ARIMA(ref['Value'], order=arima_order).fit()

    # Print the summary
    print(arima_model_fit.summary())

    # Forecasting
    arima_forecast = arima_model_fit.get_forecast(steps=10)
    arima_forecast_mean = arima_forecast.predicted_mean

    # Plotting the original series and forecasted values
    plt.figure(figsize=(12, 6))
    plt.plot(ref['Value'], label='Train')
    plt.plot(arima_forecast_mean.index, arima_forecast_mean, label='ARIMA Forecast')
    plt.title('ARIMA Model Forecast vs Actual')
    plt.legend()
    plt.show()

    forecast_steps = 10
    forecast =  arima_model_fit.forecast(steps=forecast_steps)
    print(forecast)

# Assuming df_dict is a dictionary containing the time series data

```

```python
from IPython.display import display

for key in df_dict:
    display(f"Running ARIMA for {key}")
    arima_model(df_dict[key])
```
