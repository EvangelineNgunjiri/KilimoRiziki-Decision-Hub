# EDA

```python
# Summary statistics of the data
print("Summary statistics of the data:")
print(rain_df.describe())
```

```python
# Correlation matrix
rain_df_copy=(rain_df.copy()).drop(columns ='Region')
print("Correlation matrix:")
correlation_matrix = rain_df_copy.corr()
print(correlation_matrix)

# Heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# Boxplot of rainfall for each region
plt.figure(figsize=(12, 8))
sns.boxplot(data=rain_df.melt(id_vars='Region', var_name='Year', value_name='Rainfall'), x='Region', y='Rainfall')
plt.title('Boxplot of Rainfall for Each Region')
plt.show()
```

```python
# Line plot of rainfall for each region over the years
plt.figure(figsize=(12, 8))
for region in rain_df['Region']:
    sns.lineplot(data=rain_df.set_index('Region').T, x=rain_df.columns[1:], y=region, label=region)

plt.title('Rainfall Over the Years for Each Region')
plt.xlabel('Year')
plt.ylabel('Rainfall (mm)')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```

# Modeling

```python

from sklearn.linear_model import LinearRegression

# Extract years and transpose the DataFrames
years = [2023, 2024, 2025, 2026, 2027, 2028]

# Function to predict rainfall
def predict_rainfall(df, years):
    df = df.set_index('Region')  # Set 'Region' as index
    df = df.dropna(axis=1)  # Drop columns with NaN values

    X = np.array(df.columns, dtype=int).reshape(-1, 1)

    predictions = {}
    for region in df.index:
        y = np.array(df.loc[region], dtype=float).reshape(-1, 1)

        model = LinearRegression()
        model.fit(X, y)

        future_years = np.array(years).reshape(-1, 1)
        future_predictions = model.predict(future_years)

        predictions[region] = future_predictions.flatten()

    return predictions

# Predict future rainfall
rainfall_predictions = predict_rainfall(rain_df, years)

rainfall_predictions_df = pd.DataFrame(rainfall_predictions, index=years)
```

## Model Evaluation


```python
from sklearn.metrics import mean_absolute_error

# Function to evaluate the model's performance for rainfall
def evaluate_model_rainfall(df, years):
    df = df.set_index('Region')

    X = np.array(df.columns, dtype=int).reshape(-1, 1)

    mae_values = []
    for region in df.index:
        y_true = np.array(df.loc[region], dtype=float).reshape(-1, 1)

        model = LinearRegression()
        model.fit(X, y_true)

        y_pred = model.predict(X)

        # Evaluate performance using MAE
        mae = mean_absolute_error(y_true, y_pred)
        mae_values.append(mae)

    return mae_values

# Evaluate the model for rainfall
rainfall_mae_values = evaluate_model_rainfall(rain_df, years)

# Display the evaluation results
print("\nRainfall MAE for each region:")
print(rainfall_mae_values)

# Calculate average MAE for rainfall
avg_rainfall_mae = np.mean(rainfall_mae_values)
print("\nAverage Rainfall MAE:", avg_rainfall_mae)
```
